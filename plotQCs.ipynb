```python
from qiskit import *
from qiskit.circuit import ParameterVector
qubit_count = 16
ppq = 3
kernel = QuantumCircuit(qubit_count)
thetas = ParameterVector('θ', qubit_count*ppq)

#Start
for i in range(qubit_count):
    kernel.rx(thetas[3*i], i)
    if i != qubit_count - 1:
        kernel.crx(thetas[3*i + 1], i, i+1)
    else:
        kernel.crx(thetas[3*i + 1], i, 0)
    
    kernel.ry(thetas[3*i + 2], i)
#End

kernel.draw(output='mpl')
```
